```python
#Строго обоснуйте метод rejection sampling. Смоделируйте выборку размера 100 из распределения, заданного ниже, используя генератор равномерного распределения.


def func(x):
    return 1 / ((1 + x*x) * np.pi)

count = 0
randoms = []
n = 10
while count < 10000:
    x = rng() * 2*n - n
    y = rng()/3  
    if y <= func(x):
        randoms.append(x)
        count+=1
X = np.linspace(-n, n, 1000)
plt.hist(randoms, 50, density = 1)
plt.plot(X, 1/(1 + X*X)/np.pi)
plt.show() 
```

```python
#MCMC


import numpy as np

np.random.seed(123)
data = np.random.randn(20)

plt.hist(data)
plt.title('Histogram of observed data')
plt.xlabel('x'), plt.ylabel('# observations')
plt.show()
```

```python
from scipy.stats import norm

mu_prior_mu = 0
mu_prior_sd = 1

def calc_posterior_analytical(data, x, mu_0, sigma_0):
    sigma = 1.
    n = len(data)
    mu_post = (mu_0 / sigma_0**2 + data.sum() / sigma**2) / (1. / sigma_0**2 + n / sigma**2)
    sigma_post = (1. / sigma_0**2 + n / sigma**2)**-1
    return norm(mu_post, np.sqrt(sigma_post)).pdf(x)

ax = plt.subplot()
x = np.linspace(-1, 1, 100)
posterior_analytical = calc_posterior_analytical(data, x, mu_prior_mu, mu_prior_sd)
plt.plot(x, posterior_analytical)
plt.xlabel('mu'), plt.ylabel('belief'), plt.title('Analytical posterior')
plt.show()
```

```python
def mcmc(data, k = 10000, mu_start = 3, proposal_width = .01):
    
    mu_current = mu_start
    m = np.empty(k)
    for i in range(k):
        mu_proposal = norm(mu_current, proposal_width).rvs()

        likelihood_current = norm(mu_current, 1).pdf(data).prod()
        likelihood_proposal = norm(mu_proposal, 1).pdf(data).prod()

        # Compute prior probability of current and proposed mu
        prior_current = norm(mu_prior_mu, mu_prior_sd).pdf(mu_current)
        prior_proposal = norm(mu_prior_mu, mu_prior_sd).pdf(mu_proposal)

        p_current = likelihood_current * prior_current
        p_proposal = likelihood_proposal * prior_proposal

        p_accept = min(1, p_proposal / p_current)

        accept = np.random.rand() < p_accept

        if accept:
            mu_current = mu_proposal
        m[i] = mu_current
        
    return m
        
sample = mcmc(data)
plt.plot(x, posterior_analytical)
plt.hist(sample, density = True, bins = 50)
plt.xlabel('mu'), plt.ylabel('belief')
plt.show()

    
   
```

```python
sample1 = mcmc(data, proposal_width = .001)
sample2 = mcmc(data, proposal_width = .1)
sample3 = mcmc(data, proposal_width = 10)
sample4 = mcmc(data, proposal_width = 100)

plt.figure(1, figsize = (10, 20))

plt.subplot(411)
plt.plot(sample1)
plt.xlabel('Samples'), plt.ylabel('Resulting Values of mu')
plt.title('Proposal Width is 0.001') 

plt.subplot(412)
plt.plot(sample2)
plt.xlabel('Samples'), plt.ylabel('Resulting Values of mu')
plt.title('Proposal Width is 0.1') 

plt.subplot(413)
plt.plot(sample3)
plt.xlabel('Samples'), plt.ylabel('Resulting Values of mu')
plt.title('Proposal Width is 10')

plt.subplot(414)
plt.plot(sample4)
plt.xlabel('Samples'), plt.ylabel('Resulting Values of mu')
plt.title('Proposal Width is 100') 
plt.show()
```
